```python
from scipy import misc
```

```python
# Create a random image
img = misc.ascent()
```

```python
img.shape
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
# Create a filter of 3x3
f = numpy.array([
    [1,2,-1],
    [0,0,1],
    [-1,-2,-1]
])
```

```python
f.shape
```

```python
from scipy.signal import convolve2d
```

```python
# Convolution using scipy 
cimg = convolve2d(img, f)
```

```python
cimg
```

```python
plt.imshow(cimg)
```
